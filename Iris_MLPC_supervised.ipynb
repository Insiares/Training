```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.neural_network import MLPClassifier
import seaborn as sns
sns.set_theme()
```

```python
fleurs = datasets.load_iris()

X = fleurs.data
y = fleurs.target

X.shape #normalement m=150 n=4
```

Modele r√©seau

```python
model = MLPClassifier(hidden_layer_sizes=(10, 10, 10), max_iter=1000)
model.fit(X, y)
model.score(X, y)
```

```python

colormap= np.array(['Red','green','blue'])
fig= plt.figure()
plt.rcParams['figure.figsize']= [35,25]
plt.rcParams['font.weight']="bold"
plt.subplots_adjust(wspace=0.2, hspace=0.2)
fig.suptitle('Iris dataset clustering', fontsize=25,fontweight="bold", color="black",
             position=(0.5,1.01))
ax1 = fig.add_subplot(221)
ax1.scatter(X[:,3],X[:,1], c=colormap[y])
ax1.set_title('Target', color="green")

ax2 = fig.add_subplot(222)
ax2.scatter(X[:,3],X[:,1], c=colormap[model.predict(X)])
ax2.set_title('Model', color="orange")
#verbeux
```
